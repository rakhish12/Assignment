```python
# Sum of first 10 natural numbers

sum = 0

for i in range(1, 11):
    sum = sum + i

print("Sum =", sum)
```

```python
# Factorial of a number

num = int(input("Enter a number: "))

fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial =", fact)
```

```python
# Fibonacci Series

n = int(input("How many terms? "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    
    c = a + b
    a = b
    b = c
```

```python
# Student Result System

# Student details
name = input("Enter Student Name: ")
roll_no = input("Enter Roll Number: ")

# Marks input
marks = 0

for i in range(1, 6):
    subject_marks = float(input("Enter marks of Subject " + str(i) + ": "))
    marks = marks + subject_marks

# Percentage calculation
percentage = marks / 5

# Grade calculation
if percentage >= 90:
    grade = "A+"

elif percentage >= 80:
    grade = "A"

elif percentage >= 70:
    grade = "B"

elif percentage >= 60:
    grade = "C"

elif percentage >= 40:
    grade = "D"

else:
    grade = "Fail"

# Result display
print("\n----- STUDENT RESULT -----")
print("Name :", name)
print("Roll No. :", roll_no)
print("Percentage :", percentage)
print("Grade :", grade)
```
